```python
import random

import finesse
from finesse.analysis.actions import Action
from finesse.analysis.actions.base import AnalysisState


class TamperMonkey(Action):
    def __init__(self, name="TamperMonkey", verbose: bool = False):
        super().__init__(name)
        self.verbose = verbose

    def _requests(self, model: finesse.Model, memo: dict, first: bool = True):
        self.chosen_parameter = random.choice(
            [par for par in model.all_parameters if par.changeable_during_simulation]
        )
        memo["changing_parameters"].append(self.chosen_parameter)

    def _do(self, state: AnalysisState):
        old_val = self.chosen_parameter.value
        self.chosen_parameter.value += 1
        if self.verbose:
            print(
                f"Changing parameter '{self.chosen_parameter.full_name}' from {old_val} to {self.chosen_parameter.value}"
            )


model = finesse.Model("l l1 P=1")
sol = model.run(TamperMonkey(verbose=True))
```
